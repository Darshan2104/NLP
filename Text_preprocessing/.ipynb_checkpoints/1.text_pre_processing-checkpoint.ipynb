````python
# import the necessary libraries
import nltk
import string
import re
````

## 1. lower the text

````python
def lower_text(input_string):
    output_string = input_string.lower()
    return output_string
input = "Hey, did you know that the summer break is coming? Amazing right !! It's only 5 more days !!"
lower_text(input)
````

## 2. remove numbers or convert it into text

````python
# remove Numbers
def remove_number(input):
    output = re.sub(f'\d+','',input)
    return output


# convert numbers in to text
import inflect
p = inflect.engine()
def number2text(input):
    output = ""
    temp_str = input.split()
    new_String = []
    for word in temp_str:
        if word.isdigit():
            temp = p.number_to_words(word)
            new_String.append(temp)
        else:
            new_String.append(word)
    output = ' '.join(new_String)
    return output

input = "There are 3 balls in this bag, and 12 in the other one."
print("Input : ",input,end="\n")
print("Remove number : "+remove_number(input))
print("Replace number with text : ",number2text(input))
````

## 3. Remove punctuations

````python
# if we will not remove [been. been, been!] these all 3 are treated differently
def remove_punctuation(input):
    # str.maketrans(A,B,C) ===> replace A with B and remove C from the string
    translator = str.maketrans('', '', string.punctuation)
    return input.translate(translator)
    
input = "been. into been! !~` @ # $ % ^ & * ( )  \ _ + = - { [  \" \' : ; < , > . ? / |"
remove_punctuation(input)

# Why we are not using regular expression because, we we have to seperate punctuation from word it self i.e been!!! , been? so it is better to go char level
````

## 4. remove extra white space

````python
def remove_extra_white_space(input):
    print(input.split())
    output = " ".join(input.split())
    return output

input = "   we don't need   the given questions"
remove_extra_white_space(input)
````

## 5. Remove default stop words

````python
nltk.download('stopwords')
nltk.download('punkt')
````

````python

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
# for indic language use 'inltk' library

def remove_stop_words(input):
    
    stop_words = set(stopwords.words("english"))
    # print(len(stop_words))
    # print(stop_words)

    #  print((word_tokenize(input)))
    tokenized = word_tokenize(input)
    filtered_text = [word for word in tokenized if word not in stop_words]
    # print(filtered_text)

    return filtered_text

input = "This is a sample sentence and we are going to remove the stopwords from this. and I'm here to tell you that, I've not written this sentence"
remove_stop_words(input)
````

## 6. Stemming
https://www.geeksforgeeks.org/introduction-to-stemming/

````python
from nltk.stem.porter import PorterStemmer
from nltk.stem.snowball import SnowballStemmer
from nltk.stem.lancaster import LancasterStemmer
from nltk.tokenize import word_tokenize

def Porter_Stemmer(input_text):
    stemmer = PorterStemmer()
    tokenized = word_tokenize(input_text)
    stems = [stemmer.stem(word) for word in tokenized]
    return stems

def Snowball_Stemmer(input_text):
    # it is also called multilingual stemmer
    stemmer = SnowballStemmer(language='english')
    tokenized = word_tokenize(input_text)
    stems = [stemmer.stem(word) for word in tokenized]
    return stems

def Lancaster_Stemmer(input_text):
    stemmer = LancasterStemmer()
    tokenized = word_tokenize(input_text)
    stems = [stemmer.stem(word) for word in tokenized]
    return stems

# input = "data science uses scientific methods algorithms and many types of processes"
input = "cared university fairly easily singing sings sung singer sportingly"
print("=========== Porter_Stemmer ==========")
print(Porter_Stemmer(input))
print("=========== Snowball_Stemmer ==========")
print(Snowball_Stemmer(input))
print("=========== Lancaster_Stemmer ==========")
print(Lancaster_Stemmer(input))
````

## 7. lemmatization

````python
nltk.download('wordnet')
````

````python
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

def lemmatiaztion(input):
    lemmatizer = WordNetLemmatizer()
    tokenized = word_tokenize(input)
    # v is for verb. We can use N for noun
    ans = [lemmatizer.lemmatize(word, pos ='v') for word in tokenized]
    return ans

input = "data science uses scientific methods algorithms and many types of processes. I saw something unforgotable thing which can not even come to thought"
lemmatiaztion(input)
````

## 2. Chunking

````python
from nltk.tokenize import word_tokenize
from nltk import pos_tag

def chunking(input_text, grammer):
    tokenized = word_tokenize(input_text)
    word_pos = pos_tag(tokenized)
    chunkParser = nltk.RegexpParser(grammer)

    tree = chunkParser.parse(word_pos)
    for subtree in tree.subtrees():
        print(subtree)
    # tree.draw()

input_text = "the little yellow bird is flying in the sky"
grammer = "NP: {<DT>?<JJ>*<NN>}"
chunking(input_text, grammer)
````

````python
# To extract patterns like "A as in Apple," "B like Ballon," and so on, purely based on dependency parsing, you can define grammatical rules that capture the specific dependency relationships between the alphabet, the preposition, and the corresponding word. Here's an example of how you can create such rules using spaCy:

# ```python
import spacy
from spacy.matcher import Matcher

# Example user prompt
prompt = "My first character in the pan card is A, A as in Apple."

# Load the English language model in spaCy
nlp = spacy.load("en_core_web_sm")

# Define the grammatical patterns
grammatical_patterns = {
    "alphabet_preposition_word": {
        "pattern": [{"DEP": "nsubj"}, {"DEP": "prep"}, {"DEP": "pobj"}],
        "entities": [(1, 2, "PREPOSITION")]
    }
}

# Perform dependency parsing and extract the desired patterns
doc = nlp(prompt)

extracted_phrases = []
highlighted_words = []

for sent in doc.sents:
    for pattern in grammatical_patterns.values():
        matcher = Matcher(nlp.vocab)
        print(pattern["pattern"])
        matcher.add("Pattern", [pattern["pattern"]])

        matches = matcher(sent)
        for match_id, start, end in matches:
            span = sent[start:end]
            phrase = " ".join(token.text for token in span)

            extracted_phrases.append(phrase)

            # Extract the highlighted word
            entities = pattern.get("entities")
            if entities:
                entity_start, entity_end, entity_label = entities[0]
                highlighted_word = span[entity_start:entity_end].text
                highlighted_words.append((highlighted_word, entity_label))

# Print the extracted phrases and highlighted words
print("Extracted Phrases:", extracted_phrases)
print("Highlighted Words:", highlighted_words)
# ```

# In this code snippet, we define the `grammatical_patterns` dictionary, which contains the grammatical patterns for extracting the desired phrases. Each pattern consists of a list of dictionaries, where each dictionary specifies a dependency relationship between tokens based on their dependency labels (e.g., "nsubj" for the subject, "prep" for preposition, "pobj" for prepositional object).

# The code then performs dependency parsing on the user prompt using spaCy and iterates through the sentences. For each sentence, it applies the defined grammatical patterns using spaCy's `Matcher`. When a match is found, it extracts the phrase based on the matched tokens.

# Additionally, if the pattern has an associated entity label, it extracts the highlighted word within that pattern and adds it to the `highlighted_words` list along with the corresponding entity label.

# When you run the code, it will output:

# ```
# Extracted Phrases: ['A as in Apple']
# Highlighted Words: [('as', 'PREPOSITION')]
# ```

# This demonstrates the extraction of the phrase "A as in Apple" based on the dependency parsing rules, and the highlighted word "as" is recognized as a preposition.

# Feel free to modify the code to suit your specific requirements or expand the `grammatical_patterns` dictionary to include additional patterns for other alphabets and phrases.
````

````python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

# sentence = "In the alphabet, A is for Apple, B stands for banana, and B for Bombay."
sentence = "My PAN number starts with a A is for america"

# Tokenize the sentence into words
tokens = word_tokenize(sentence)

# Perform POS tagging
pos_tags = pos_tag(tokens)

# Initialize variables
phrases = []
current_phrase = []

# Iterate over POS tags
for word, pos in pos_tags:
    print("\n")
    print(f"{word}----> {pos}")
    print(f"current_phrase -----> {current_phrase}")
    print(f"phrases -----> {phrases}")
    print("\n")
    if pos == 'VBZ' or pos == 'VBD':
        # Found a verb indicating "is" or "stands"
        current_phrase.append(word)
    elif pos == 'IN' or pos == 'CC':
        # Found a preposition or coordinating conjunction
        current_phrase.append(word)
        phrases.append(" ".join(current_phrase))
        current_phrase = []
    elif pos == 'NNP' or pos == 'NN':
        # Found a noun
        current_phrase.append(word)

# Check if there is a remaining phrase
if current_phrase:
    phrases.append(" ".join(current_phrase))

# Print the extracted phrases
print("\n\n\n\n")
for phrase in phrases:
    print(phrase)
````

````python
import spacy
from spacy import displacy

# Load the language model
nlp = spacy.load("en_core_web_sm")

# sentence = 'First charector of my PANcard is A is for Apple'
# sentence = 'First charector of my PANcard is B stands for Banana'
# nlp function returns an object with individual token information,
# linguistic features and relationships
# Deep Parsing
def get_DP(sentences):
    for sentence in sentences:
        print("\n\n\n")
        print(f"input text : {sentence}\n")
        doc = nlp(sentence)

        print("{:<15} | {:<8} | {:<15} | {:<20}".format(
            'Token', 'Relation', 'Head', 'Children'))
        print("-" * 70)

        for token in doc:
            # Print the token, dependency nature, head and all dependents of the token
            print("{:<15} | {:<8} | {:<15} | {:<20}"
                .format(str(token.text), str(token.dep_), str(token.head.text), str([child for child in token.children])))

        # Use displayCy to visualize the dependency
        displacy.render(doc, style='dep', jupyter=True, options={'distance': 120})
        print("\n")
````

````python
basic_sentences = [
    "First character of my PAN card is A stands for Apple.",
    "First character of my PAN card is C stands for Cat.",
    "First character of my PAN card is D stands for Dog.",
    "First character of my PAN card is E stands for Elephant.",
    "First character of my PAN card is F stands for Fox."
]
get_DP(basic_sentences)
````
